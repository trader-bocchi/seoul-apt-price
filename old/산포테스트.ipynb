```python
import time, random
import requests
from playwright.sync_api import sync_playwright, TimeoutError as PWTimeoutError

API_URL = (
    "https://new.land.naver.com/api/articles/complex/118771"
    "?realEstateType=APT%3APRE%3AABYG%3AJGC"
    "&tradeType="
    "&tag=%3A%3A%3A%3A%3A%3A%3A%3A"
    "&rentPriceMin=0&rentPriceMax=900000000"
    "&priceMin=0&priceMax=900000000"
    "&areaMin=0&areaMax=900000000"
    "&oldBuildYears&recentlyBuildYears"
    "&minHouseHoldCount&maxHouseHoldCount"
    "&showArticle=false"
    "&sameAddressGroup=false"
    "&minMaintenanceCost&maxMaintenanceCost"
    "&priceType=RETAIL"
    "&directions="
    "&page=1"
    "&complexNo=118771"
    "&buildingNos=&areaNos="
    "&type=list&order=rank"
)

COMPLEX_URL = "https://new.land.naver.com/complexes/118771"


def requests_with_backoff(url, headers, cookies, max_retry=6):
    s = requests.Session()
    for i in range(max_retry):
        r = s.get(url, headers=headers, cookies=cookies, timeout=20)
        # 네이버가 429 대신 JSON으로 TOO_MANY_REQUESTS를 줄 때도 있음
        is_429 = (r.status_code == 429)

        try:
            data = r.json()
        except Exception:
            data = None

        is_rate_limited = is_429 or (isinstance(data, dict) and data.get("code") == "TOO_MANY_REQUESTS")

        if not is_rate_limited:
            return r, data

        # 지수 백오프 + 랜덤 지연(사람처럼)
        sleep_s = (2 ** i) + random.uniform(0.5, 2.5)
        print(f"[rate limited] retry {i+1}/{max_retry}, sleep {sleep_s:.1f}s")
        time.sleep(sleep_s)

    return r, data


def fetch_articles_via_playwright():
    with sync_playwright() as p:
        browser = p.chromium.launch(headless=True)  # 디버깅은 False로 바꾸면 됨
        context = browser.new_context(locale="ko-KR")
        page = context.new_page()

        # 1) 단지 페이지 진입
        page.goto(COMPLEX_URL, wait_until="domcontentloaded")
        page.wait_for_timeout(2500)

        # 2) 페이지가 내부적으로 보내는 API 요청을 기다려서 Authorization 뽑기
        #    (가장 확실: 실제 네트워크 요청을 캡처)
        try:
            req = page.wait_for_request(
                lambda r: ("/api/articles/complex/118771" in r.url),
                timeout=15000
            )
        except PWTimeoutError:
            # 만약 자동으로 호출이 안 잡히면, 화면 상호작용을 조금 넣어서 트리거
            page.reload(wait_until="domcontentloaded")
            page.wait_for_timeout(2500)
            req = page.wait_for_request(
                lambda r: ("/api/articles/complex/118771" in r.url),
                timeout=15000
            )

        req_headers = req.headers
        bearer = req_headers.get("authorization")
        if not bearer:
            raise RuntimeError("Authorization(Bearer) 헤더를 못 찾았어. 네이버가 호출 구조를 바꿨을 수 있어.")

        # 3) 브라우저 쿠키 추출
        cookie_list = context.cookies()
        cookies = {c["name"]: c["value"] for c in cookie_list}

        # 4) requests에서 필요한 최소 헤더만 구성 (과도한 쿠키/헤더는 오히려 역효과)
        headers = {
            "accept": "*/*",
            "accept-language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
            "authorization": bearer,
            "referer": COMPLEX_URL,
            "user-agent": page.evaluate("() => navigator.userAgent"),
        }

        # 5) 요청 (레이트리밋이면 백오프)
        #    첫 요청은 약간 쉬고 (사람처럼)
        time.sleep(random.uniform(1.2, 2.8))
        r, data = requests_with_backoff(API_URL, headers=headers, cookies=cookies)

        browser.close()
        return r.status_code, data


if __name__ == "__main__":
    status, data = fetch_articles_via_playwright()
    print("HTTP:", status)
    print(data)
```
